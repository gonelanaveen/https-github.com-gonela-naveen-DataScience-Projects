```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.cm as cm
cars_df=pd.read_csv("cars93.csv")
cars_df.head()
```

```python
fig, (ax,ax1)=plt.subplots(1,2)
fig.set_figwidth(15)
fig.set_figheight(5)
ax.boxplot(cars_df['Horsepower'], vert=False)
ax1.boxplot(cars_df['MPG.city'], vert=False)
ax.set_title('Horsepower')
ax.set_xlabel('Power')
ax.set_ylabel('Torque')
ax1.set_title('citymileage')
ax1.set_xlabel('power')
ax1.set_ylabel('Torque')
fig.tight_layout()
```

```python
plt.scatter(cars_df['Horsepower'], cars_df['MPG.city'])
ax.set_xlabel('Horsepower')
ax.set_ylabel('MPG.City')
ax.set_title('Horsepower vs MPG.city', fontsize=10)
```

```python
#The following lines of code create a blank canvas to plot on
fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(7)
#Data is fed and plotted using the following lines
cars_df.plot(ax = ax, x = "Horsepower", y = "MPG.highway", kind = "line", )
cars_df.plot(ax = ax, x = "Horsepower",y= "MPG.city", kind = "line", linestyle='--')
```

```python
#First sort the data to get a proper line chart
cars_df=cars_df.sort_values(by="Horsepower")
cars_df.describe()
#The following lines of code create a blank canvas to plot on
fig, ax = plt.subplots()
fig.set_figwidth(10)
fig.set_figheight(4)
#Data is fed and plotted using the following lines
cars_df.plot(ax = ax, x = "Horsepower", y = "MPG.highway", kind = "line", )
cars_df.plot(ax = ax, x = "Horsepower",y= "MPG.city", kind = "line", linestyle='--')
```

```python

cars_df=cars_df.sort_values(by="Horsepower")

fig, ax= plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(10)


cars_df.plot(ax=ax, x="Horsepower", y="EngineSize", kind="line")
cars_df.plot(ax=ax, x="Horsepower", y="MPG.city", kind="line", linestyle="--", fontsize=10)
cars_df.plot(ax=ax, x="Horsepower", y="MPG.highway", kind="line")
plt.plot(cars_df["Horsepower"], [i/100 for i in cars_df["RPM"]],label="RPM")
ax.set_xlabel("Horsepower")
ax.set_ylabel("EngineSize")
plt.legend()
```

```python
fig, ax= plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(10)
ax.barh(cars_df["DriveTrain"], cars_df["MPG.city"], height=0.3, label="mileage")
```

```python
fig, ax= plt.subplots()
fig.set_figwidth(15)
fig.set_figheight(10)
ax.bar(cars_df["DriveTrain"], cars_df["MPG.city"], width=0.3, label="mileage")
```

```python
grouped_data = cars_df[["Passengers","Type"]].groupby(by= ["Passengers","Type"]).size().unstack().reset_index()

#Stacked Bar Graph can be plotted using the grouped data, as follows:
grouped_data.plot(x="Passengers",kind="bar",figsize=(10,7))
```

```python
#Error bars can be added to the stacked bar graph with the 'yerr' argument as follows
grouped_data.plot(x="Passengers",kind="bar",stacked=True,colormap=cm.seismic,figsize=(10,7),xerr=np.std(cars_df["Passengers"]))
```

```python
grouped_cars = cars_df[["MPG.city","MPG.highway","RPM","DriveTrain"]].groupby(by="DriveTrain").mean().T
```

```python
fig = plt.figure()
fig.set_figwidth(10)
fig.set_figheight(7)
grouped_cars.loc['RPM'] /= 100
width=0.2 # We assign the value of the width of the bar and on the number of groups.
ind=list(range(len(cars_df['DriveTrain'].unique()))) 
plt.bar([i for i in ind], height=grouped_cars["4WD"], label="4WD", width=width)
plt.bar([i+width for i in ind], height=grouped_cars["Front"], width=width, bottom=0, label="Front")
plt.bar([i+width*2 for i in ind], height=grouped_cars["Rear"], label="Rear", width=width, bottom=0)
plt.suptitle("Mileage in city, Mileage on highway,RPM vs DriveTrain", fontsize=16)
plt.xlabel("Mileage in city, Mileage in highway, RPM")
plt.ylabel("Average per DriveTrain type")
plt.xticks([i+width for i in ind],["Mileage in City","Mileage in Highway","RPM"])
plt.legend()
```

```python
plt.bar([i for i in ind], height=grouped_cars["4WD"], label="4WD", width=width)
plt.bar([i+width for i in ind], height=grouped_cars["Front"], width=width, bottom=0, label="Front")
plt.bar([i+width*2 for i in ind], height=grouped_cars["Rear"], label="Rear", width=width, bottom=0)
```
